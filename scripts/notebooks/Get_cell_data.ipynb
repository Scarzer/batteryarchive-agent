```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_url = "http://0.0.0.0:4000"
```

```python
def getCellTimeSeries(cell_id, test):
    df = pd.read_json(data_url + "/cells/" + cell_id + "/tests/" + test + "/ts");
    return df 

def getCellMeta(cell_id):
    df = pd.read_json(data_url + "/cells/" + cell_id + "/meta")
    return df    
```

```python
df_cells = pd.read_json(data_url + "/cells")
df_cells
```

```python
cell_id = "Arbin_Cycle"
df_md = getCellMeta(cell_id)
df_ts = getCellTimeSeries(cell_id, "cycle")
df_ts.plot(x="test_time", y=["v", "i"], figsize=(12, 8))
```

```python
cell_id = "SNL_Abuse"
df_md = getCellMeta(cell_id)
df_ts = getCellTimeSeries(cell_id, "abuse")
df_ts.plot(x="test_time", y=["v", "above_punch_temperature"], figsize=(12, 8))
```
